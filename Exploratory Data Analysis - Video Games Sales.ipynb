```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
plt.rcParams["axes.grid"] = True
plt.rcParams["grid.linestyle"] = "--"
```

```python
#loading data
data = pd.read_csv('vgsales.csv')
```

```python
data.sample(5)
```

```python
#take a look at statistical summery of data
data.describe()
```

```python
#the info function lets us see the properties of each field
data.info()
```

```python
#droping null value
data.dropna(how='any',inplace=True)
```

```python
# Make sure the data is ordered by Global_Sales
data.sort_values(by="Global_Sales", ascending=False, inplace=True)

# Reset the index
data.reset_index(drop=True, inplace=True)

# Rearrange the Rank field
data["Rank"] = np.arange(1, len(data) + 1)
```

```python
data["Year"] = data["Year"].astype(int)
```
