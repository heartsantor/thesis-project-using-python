```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('dataset.csv')
```

```python
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
x = df.drop(['Name','Publisher','Year','Merketing_Cost','Rating','Global_Sale'],axis=1)
```

```python
x
```

```python
y = df['Global_Sale']
```

```python
y
```

One Hot Encoding

```python
ctg = pd.get_dummies(x['Genre'],drop_first=True)
```

```python
ctg
```

```python
x = x.drop('Genre',axis=1)
```

```python
x
```

```python
x = pd.concat([x,ctg],axis=1)
```

```python
x
```

spriting dataset Treading set and test set


```python
from sklearn.model_selection import train_test_split
```

```python
y
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.25, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(xtrain, ytrain)
```

```python
ytest
```

```python
pred = regressor.predict(xtest)
```

```python
pred
```

```python
regressor.score(xtest,ytest)
```

# R-squared value

```python
from sklearn.metrics import r2_score
```

```python
score =r2_score(ytest,pred)
```
